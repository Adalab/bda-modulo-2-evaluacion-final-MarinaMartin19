```python
# 1. Conseguimos todos los generos sin duplicados 

generos = set()  # un set no permite duplicados.

for peli in datos_pelis:

    generos.add(peli['genero']) # Anadimos cada genero

print(generos)
```

```python
# Insertamos cada género en MySQL: 



for genero in generos:


    query = 'INSERT INTO genres (name) VALUES (%s)'

    mycursor.execute(query, (genero,))

    cnx.commit()
print ('Generos insertados correctamente')
```

```python
for peli in datos_pelis:

    id_api = peli['id']

    titulo = peli ['titulo']

    año = peli['año']
    duracion = peli ['duracion']
    adultos = peli['adultos']
    genero = peli ['genero']


    # Buscamos el id del genero en SQL

    query = 'SELECT id FROM genres WHERE name = %s'

    mycursor.execute(query, (genero,))
    genre_id = mycursor.fetchone()[0]

    # insertamos la pelicula 

    query = '''
     INSERT INTO movies (id_api, title, year, length, adult, genre_id)
     VALUES (%s, %s, %s, %s, %s, %s)'''

    values = (id_api, titulo, año, duracion, adultos, genre_id)


    mycursor.execute(query, values)
    cnx.commit()

print("Películas insertadas correctamente")
```

```python
for peli in datos_pelis:

    #Insertamos la pelicula otra vez solo para obtener su id

    id_api = peli ['id']
    query = 'SELECT id FROM movies WHERE id_api = %s'

    mycursor.execute(query, (id_api,))
    movie_id = mycursor.fetchone()[0]

    for subt in peli ['subtitulos']:
        query = 'INSERT INTO movie_subtitles (movie_id, subtitle) VALUES (%s, %s)'
        mycursor.execute(query, (movie_id, subt))
        cnx.commit()

print('Subtitulos insertados correctamente')
```
